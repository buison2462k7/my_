Bài 3 : Tính tam giác: Nhâp vào ba số a,b,c bất kì. Kiểmtra xem nó có thể làđộ dài ba cạnh hay không, nếu không thì in ra màn hình 'Khong la ba canh cua mot tam giac'. Ngược lại, kiểm tra xem nó là tam giác gì: Đều, Cân, Vuông, hay Tam giác thường. Tính diện tích, chu vi của tam giác và in ra mànhình.

```python
a,b,c=map(float,input().split())
if a+b<=c or a+c<=b or c+b<=a:
    print(f"{a},{b},{c} khong phai 3 canh tam giac")
elif a==b==c:
    print("La tam giac deu")
    p=(a+b+c)/2
    Dien_tich=((p-a)*(p-b)*(p-c)*p)**(1/2)
    print(Dien_tich)
elif  (round(a**2+b**2,1)==round(c**2,1) and a==b) or (a**2+c**2==b**2 and a==c) or (b**2+c**2==a**2 and b==c):
    print("La tam giac vuong can")
    p=(a+b+c)/2
    Dien_tich=((p-a)*(p-b)*(p-c)*p)**(1/2)
    print(Dien_tich)
elif a==b or a==c or b==c:
    print("la tam giac can")
    p=(a+b+c)/2
    Dien_tich=((p-a)*(p-b)*(p-c)*p)**(1/2)
    print(Dien_tich)
elif a**2+b**2==c**2 or a**2+c**2==b**2 or b**2+c**2==a**2:
    print("La tam giac vuong")
    p=(a+b+c)/2
    Dien_tich=((p-a)*(p-b)*(p-c)*p)**(1/2)
    print(Dien_tich)
else:
    print("la tam giac thuong")
    p=(a+b+c)/2
    Dien_tich=((p-a)*(p-b)*(p-c)*p)**(1/2)
    print(Dien_tich)
```

Bài 5: Viết chương trình giải phương trình bậc 2 (ax2 + bx + c = 0)


```python
a,b,c=map(float,input().split())
Delta=(b**2)-(4*a*c)
if a!=0:
     print("Day la PT bac 2")
     if Delta>0:
         x1=(-b+Delta**(1/2))/2*a
         x2=(-b-Delta**(1/2))/2*a
         print(f"PT co 2 nghiem phan biet X1={x1}va x2={x2}")
     elif Delta==0:
         x1=x2=-b/(2*a)
         print(f'PT co nghiem kep x1=x2={x1}')
     else:
         print("PT vo nghiem")
else:
    print("PT ko phai bac 2")



```
