W5A16 Viết hàm kiểm tra một số có phải số Amstronhg hay không? Hàm trả về kiểu bool, có tham số là một số
nguyên x

```python
def Amstronhg(n):
    t=len(str(n))
    s=0
    a=n
    x=0
    if n<0:
        return False
    while a!=0:
        x=a%10
        s+=x**t
        a//=10
    return s==n
n=int(input())
print(Amstronhg(n))
        
```

W5A17 Viết hàm kiêểm tra một số có phải số chính phương hay không? Hàm trả về kiểu bool, có tham số là một số
nguyên x

```python
def chinh_phuong(n):
    if n<0:
        return False
    x=int(n**0.5)
    if x*x==n:
        return True
    return False
n=int(input())
print(chinh_phuong(n))
```

W5A21 Viết hàm is_one_product_of_others(a, b, c) trả về True nếu một trong ba số bằng tích của hai số còn lại

```python
def is_one_product_of_others(a,b,c):
    return a*b==c or a*c==b or c*b==a
a,b,c=map(int,input().split())
print(is_one_product_of_others(a,b,c))
```

W5A27 Viết hàm trả về số lượng chữ số khác 0 trong n số nguyên nhập vào từ bàn phím (một list)

```python
def day_so(n):
    t=0
    for i in range(len(str(n))):
        if n[i]!=0:
            t+=1
    return t
n=list(map(int,input()))
print(day_so(n))
```
