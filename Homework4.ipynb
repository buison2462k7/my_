W4A16 In các số chẵn chia hết cho 3 bé hơn 100

```python
for i in range(0,100):
    if i%3==0:
        print(i)
```

W4A17 Nhập vào số nguyên dương a, in ra bảng cửu chương của a 


```python
a=int(input())
for i in range(1,11):
    b=a*i
    print(f'{a}*{i}={b}')
```

W4A18 Nhập vào số nguyên dương a và b, in toàn bộ ước chung của a và b 


```python
a,b=map(int,input().split())
c=(a+b)/2
for i in range(1,int(c)):
    if a%i==0 and b%i==0:
        print(i,end=' ')
```

dương chẵn nhỏ hơn hoặc bằng n (n <= 1000)

```python
n=int(input())
for i in range(1,n+1):
    if i%2==0:
        print(i,end=' ')
```

W4A20 Nhập vào một số nguyên dương n, kiểm tra xem n có phải là số dạng 2^k hay không

```python
n=int(input())
for i in range(1,int(n**(1/2))+1):
    if i*i==n:
        print('co')
else:
     print('no')
```
