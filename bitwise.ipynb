Kiểm tra chẵn lẻ

```python
n=int(input())
if n&1==0:
    print('Chẵn')
else:
    print('Lẻ')
```

```python
n = int(input())
x = list(map(int, input().split()))
for i in range(0,n):
  if x[i]==42:
    print("I've found the meaning of life!")
    break
else:
    print("It's a joke!")
    
```

```python
n=int(input())
a=0
while n!=0:
  b=n%10
  a=a*10+b
  n=n//10
print(a)
```

```python
n=int(input())
tich=1
for i in range(1,n+1):
  tich*=i
print(tich)
```
