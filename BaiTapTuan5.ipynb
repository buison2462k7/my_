```python

def swap(a,b):
    a,b=b,a
    return a,b
x,y=swap(15,9)
print(x,y)
```

W5A3 Viết hàm kiểm tra một số có phải là số nguyên tố hay không? 

```python
def dau_vao(x):
    if x<2:
        return ('Ko la so nguyen to')
    for i in range(2,x):
        if x%i==0:
            return ('Ko la so nguyen to')
    return ('La so nguyen to')
print(dau_vao(1))
```

W5A4  VIết hàm kiểm tra một số có phải số hoàn hảo hay không, xử lý trên dữ liệu kiểu số nguyên. 


```python
def sohoanhao(x):
    if x<2:
        return 'Ko phai so hoan hao'
    t=0
    for i in range(1,x):
        if x%i==0:
            t+=i
    if t==x:
        return 'So hoan hao'
    return 'Ko phai so hoan hao'
print(sohoanhao(28))
```

W5A5 Viết hàm đưa vào một list số nguyên và một số nguyên dương k. Hãy tìm và trả về vị trí của phần tử
đầu tiên có giá trị k trong list số nguyên, nếu không có thì trả về -1 


```python
def find_k(lst, k):
    for i in range(len(lst)):
        if lst[i]==k:
            return i
    return -1
print(find_k((1,3,34,3,43,2), 43))
```

W5A6 Viết hàm tính giai thừa. Nhập vào số nguyên n, trả về giai thừa của n (n!)

```python
def giai_thua(n):
    tich=1
    for i in range(2,n+1):
        tich*=i
    return tich
print(giai_thua(5))
```

W5A9 Viết hàm trả về tổng các chữ số của số nguyên dương n 


```python
def tong_chuso(n):
    s=0
    while n!=0:
        i=n%10
        s+=i
        n//=10
    return s
print(tong_chuso(2312))

```

W5A11 Viết hàm is_product_divisible_by_10(a, b) trả về True nếu tích a*b chia hết cho 10, ngược lại False.

```python
def is_product_divisible_by_10(a, b):
    if (a*b)%10==0:
        return True
    return False
print(is_product_divisible_by_10(42,5))
```

W5A12 Viết hàm divide_with_remainder(dividend, divisor) trả về bộ (q, r) sao cho dividend = divisor*q + r và
0 <= |r| < |divisor|. 


```python
def divide_with_remainder(dividend,divisor):
    if divisor==0:
        return 'Ko hop le'
    q=dividend//divisor
    r=dividend%divisor
    return q,r
print(divide_with_remainder(-10,3))
```

W5A14  Viết hàm compute(a, b, op) thực hiện phép tính a (+|-|*|/) b tùy theo ký tự op. 


```python
def compute(a,b, op):
    if op=='+':
        return a + b
    elif op=='-':
        return a-b
    elif op=='*':
        return a*b
    elif op=='/':
        if b==0:
            return 'Ko hop le'
        return a/b
print(compute(5,6,'/'))
```

W5A15 Viết hàm is_right_triangle(a, b, c) trả về True nếu ba số có thể là ba cạnh của tam giác vuông.

```python
def is_right_triangle(a,b,c):
    if a+b<=c or a+c<=b or b+c<=a:
        return 'Ko phai tam giac'
    elif a**2 + b**2 == c**2 or a**2 + c**2 == b**2 or c**2 + b**2 == a**2:
        return 'Tam giac vuong'
print(is_right_triangle(3,4,5))
    
```
